````python
scaler_1 = StandardScaler()
scaler_1.fit(df)
type(scaler_1)
````

````python
col_names=list(df)
X_1 = pd.DataFrame(scaler_1.transform(df), columns=col_names)
````

````python
def elbow_method(dataframe, max_K):

    meanDispersions = []
    for i in max_K:
        kmeans_object = KMeans(n_clusters=i) 
        kmeans_object.fit(dataframe) 
        inertia = kmeans_object.inertia_

        meanDispersions.append(inertia)
    return meanDispersions
````

````python
K = range(1, 10)
dispersions = elbow_method(X_1, K)
````

````python
p = sns.lineplot(x=K, y=dispersions, color='blue', markers=True, dashes=False)
p.set_title("Selecting k with the Elbow Method");
````

````python
X_train_1, X_test_1 = train_test_split(X_1, test_size=0.2, random_state=42)
````

````python
model_1 = KMeans(n_clusters=4, random_state=42)
fit_1 = model_1.fit(X_test_1)
labels_km = fit_1.labels_
print( f"Er zijn dus {max(labels_km)+1} clusters" )
````

````python
s_score_km = silhouette_score(X_test_1, labels_km)
print(f"silhouette_score voor KMeans clustering is {s_score_km}")
````

````python
uitkomst_1 = fit_1.predict(X_test_1)
````
